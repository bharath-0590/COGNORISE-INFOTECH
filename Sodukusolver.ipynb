```python
def find_empty(grid):
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == 0:
                return (i,j)    #row,col
    return None

def is_valid(grid, num, pos):
    #check row
    for j in range(len(grid[0])):
        if grid[pos[0]][j] == num and pos[1] !=j:
            return False
        
    #check column 
    for i in range(len(grid)):
        if grid[i][pos[1]] == num and pos[0] != i:
            return False
        
    #check 3x3 box 
    box_x = pos[1] //3 
    box_y = pos[0] //3
    
    for i in range(box_y * 3, box_y *3 +3):
        for j in range(box_x * 3, box_x *3 +3):
            if grid[i][j] == num and (i,j) != pos:
                return False
    return True

def solve_soduku(grid):
    empty = find_empty(grid)
    if not empty:
        return True     #Puzzle Sloved
    else:
        row, col = empty
    
    for num in range(1,10):
        if is_valid(grid, num, (row,col)):
            grid[row][col] = num
            
            if solve_soduku(grid):
                return True
            
            grid[row][col] = 0
            
    return False

def print_grid(grid):
    for row in grid:
        print(" ".join(str(num) for num in row))
        
if __name__ == "__main__":
    soduku_grid =  [
        [5, 3, 0, 0, 7, 0, 0, 0, 0],
        [6, 0, 0, 1, 9, 5, 0, 0, 0],
        [0, 9, 8, 0, 0, 0, 0, 6, 0],
        [8, 0, 0, 0, 6, 0, 0, 0, 3],
        [4, 0, 0, 8, 0, 3, 0, 0, 1],
        [7, 0, 0, 0, 2, 0, 0, 0, 6],
        [0, 6, 0, 0, 0, 0, 2, 8, 0],
        [0, 0, 0, 4, 1, 9, 0, 0, 5],
        [0, 0, 0, 0, 8, 0, 0, 7, 9]
    ]
    
    if solve_soduku(soduku_grid):
        print('Soduku solved succesfully:')
        print_grid(soduku_grid)
    else:
        print('No solution exists for the soduku puzzle:')
        
```
